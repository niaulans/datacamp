```python
def raise_both(value1, value2):
    """Raise valeue1 to the power of value2 and vice verse"""
    
    new_value1 = value1 ** value2
    new_value2 = value2 ** value1
    
    new_tuple = (new_value1, new_value2)
    
    return new_tuple

raise_both(2,3)
```

```python
def shout(word1, word2):
    
    shout1 = word1 + '!!!'
    shout2 = word2 + '!!!'
    new_shout = shout1 + shout2
    
    return new_shout

yell = shout('Selamat ', ' Nia ')
print(yell)
```

```python
def shout_all(word1, word2):
    
    shout1 = word1 + '!!!'
    shout2 = word2 + '!!!'
    shout_words = (shout1, shout2)
    
    return shout_words

yell1, yell2 = shout_all('Selamat', 'Nia')
```

```python
print(yell1)
```

```python
#Tanpa function
import pandas as pd

df = pd.read_csv("tweets.csv")

langs_count = {}
col = df['lang']

for entry in col:
    if entry in langs_count.keys():
        langs_count[entry] += 1
    else :
        langs_count[entry] = 1
        
print(langs_count)
```

```python
#Dengan function
import pandas as pd 

tweets_df = pd.read_csv('tweets.csv')

def count_countries(df, col_name):
    """Return a dict with counts of occurrences as value for each key."""
    
    langs_count = {}
    col = df[col_name]
    
    for entry in col:
        if entry in langs_count.keys():
            langs_count[entry] += 1
        else:
            langs_count[entry] = 1
    
    return langs_count

result = count_countries(tweets_df, 'lang')
print(result)
```

## Scope and user-defined functions

- Tidak semua object dapat diakses di keseluruhan script
- Scope = bagian program dimana sebuah object dan variabel dapat diakses
    - Global scope : dedifinisikan dalam body script
    - Local scope : didefinisikan di dalam function
    - Built-in scope : didefinisikan di dalam pre-defined built-ins module

### Global vs local scope

```python
def square(value):
    """Return the square fof a number"""
    new_val = value ** 2 #local scope, tidak global
    return new_val

square(3)
```

```python
new_val
```

```python
new_val = 10 #Global scope

def square(value):
    """Return the square fof a number"""
    new_val = value ** 2 #local scope, tidak global
    return new_val

square(3)
```

```python
new_val
```

```python
new_val = 10 #Global scope

def square(value):
    """Return the square fof a number"""
    new_val2 = new_val ** 2 #local scope, tidak global
    return new_val2

square(3)
```

```python
new_val = 10 #Global scope

def square(value):
    """Return the square fof a number"""
    global new_val
    new_val = new_val ** 2 #local scope, tidak global
    return new_val

square(3)
```

```python
new_val
```

```python
num = 5
def func1():
    num = 3
    print(num)
func1()
```

```python
num = 5

def func2():
    global num
    double_num = num * 2
    num = 6
    print(double_num)
func2()
```

```python
num
```

```python
team = "teen titans"

def change_team():
    
    global team
    team = "justice league"

print(team)
```

```python
change_team()
team
```

### Nested Functions

```python
def outer(...):
    """..."""
    x = ...
    
    def inner(...):
        """..."""
        y = x ** 2
    return ...
```

```python
def mod2plus5(x1, x2, x3):
    """Return the remainder plus 5 of three values."""
    
    new_x1 = x1 % 2 + 5
    new_x2 = x2 % 2 + 5
    new_x3 = x3 % 2 + 5
    
    return (new_x1, new_x2. new_x3)
```

Code diatas tidak efisien jika kita ingin menjalankan komputasi berulang kali

```python
def mod2plus5(x1, x2, x3):
    """Returns the remainder plus 5 of three values."""
    
    def inner(x):
        """Returns the remainder plus 5 of a value."""
        return x % 2 + 5
    
    return (inner(x1), inner(x2), inner(x3))
```

```python
mod2plus5(1, 2, 3)
```

```python
def raise_val(n):
    """Return the inner function"""
    
    def inner(x):
        """Raise x to the power of n."""
        raised = x ** n
        return raised
    
    return inner
```

```python
square = raise_val(2) #isi parameter ini = n
cube = raise_val(3)
print(square(2), cube(2)) #isi parameter ini x
```

```python
def outer():
    """Prints the value of n."""
    n = 1
    
    def inner():
        nonlocal n
        n = 2
        print(n)
        
    inner()
    print(n)
```

```python
outer()
```

#### Scopes searched

- Local scope
- Enclosing functions
- Global 
- Built-in

```python
def three_shouts(word1, word2, word3):
    """Returns a tuple of strings concatened with !!!"""
    
    def inner(word):
        """Returns a strings concatenend with !!!"""
        return word + "!!!"
    
    return(inner(word1), inner(word2), inner(word3))
```

```python
print(three_shouts("Nia ","Ulan ", "Sari"))
```

```python
def echo(n):
    
    def inner_echo(word):
        echo_word = word * n
        return echo_word
    
    return inner_echo

twice = echo(2)
thrice = echo(3)
```
